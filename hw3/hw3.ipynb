### Упражнение 1. 
Дан список из чисел.  
Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).  

```python
from typing import Union
from math import gcd, lcm
```

```python
def find_lcm_gcd(nums: Union[list, tuple]) -> tuple:
    # lcm = НОК, gcd = НОД
    return lcm(*nums), gcd(*nums)

print(find_lcm_gcd([15, 28, 35, 239, 1305, 4, 223]))
print(find_lcm_gcd([4, 6, 12, 24]))
```

### Упражнение 2.
Даны n предложений. Определите, сколько из них содержат хотя бы одну цифру.

```python
def count_nums(sents: Union[list, tuple]) -> int:
    n = 0
    for sent in sents:
        if any(char.isdigit() for char in sent):
            n += 1
    return n

print(count_nums(["Тут нет цифр", "А тут есть цифры 123", "И тут 593458438539538е"]))
print(count_nums(["Тут нет цифр", "Совсем нет", "Да, точно нет цифр"]))
```

### Упражнение 3.
Дана строка s и символ k. Реализуйте функцию, рисующую рамку из символа k вокруг данной строки, например:  

&&&&&&&&&&  
&Текст в рамке&  
&&&&&&&&&&    

```python
def print_frame(st: str, symb: str) -> None:
    print(symb * (len(st) + 2))
    print(symb, st, symb, sep="")
    print(symb * (len(st) + 2))

print_frame("Рисуем рамочку", "&")

print() # Для разделения

print_frame("", "$")
```

### Упражнение 4.
Для введенного предложения выведите статистику символ=количество. Регистр букв не учитывается.  

```python
def count_unique(st: str):
    letters = {}
    st = st.replace(" ", "").lower()
    for i in set(st):
        letters[i] = st.count(i)
    return ", ".join([f"{k} - {v}" for k, v in letters.items()])

print(count_unique('АаА'))
print(count_unique('Голова думала'))
print(count_unique("Рамнзль мрльк-мрльк альрг-альрг"))
```

### Упражнение 5.
Используя шифр Цезаря (достаточно только букв русского алфавита, знаки препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку.  

```python
def shift(st: str, shift: int): # Только для русских букв как указано в задании, тестировал работу на windows 10 
    new_st = ""
    for i in range(len(st)):
        if 1072 <= ord(st[i]) <= 1103:
            new_st += chr((ord(st[i]) - 1072 + shift) % 32 + 1072)
        elif 1040 <= ord(st[i]) <= 1071:
            new_st += chr((ord(st[i]) - 1040 + shift) % 32 + 1040)
        else:
            new_st += st[i]
    return new_st

test = "АБВГ, $%$абвг"
print(f"Изначальная строка: {test}")
test = shift(test, -2)
print(f"Зашифрованная строка: {test}")
test = shift(test, 2)
print(f"Расшифрованная строка: {test}")
```

### Упражнение 6.
Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:  
отрицательных значений (отсортирован по убыванию);  
неотрицательных значений (отсортирован по возрастанию).  

```python
def make_lists(*args) -> tuple:
    neg, pos = [], []
    for i in args:
        if i < 0:
            neg.append(i)
        else:
            pos.append(i)
    neg = sorted(neg, reverse=True)
    pos = sorted(pos)
    return neg, pos

print(make_lists(*[1, 2, 3, 45324, -5435, 0, 0x4534, 54, -234345, -43]))
```

### Упражнение 7. 
Строка называется палиндромом, если она пишется одинаково в обоих направлениях. Например, палиндромами в английском языке являются слова «anna», «civic», «level», «hannah».  Напишите программу, запрашивающую у пользователя строку и при помощи цикла определяющую, является ли она палиндромом.  

```python
def is_palindrom(st: str) -> bool:
    return all([st[i] == st[len(st) - 1 - i] for i in range(int(len(st) / 2) + 1)])

print(is_palindrom("anna"))
print(is_palindrom("civic"))
print(is_palindrom("level"))
print(is_palindrom("hannah"))
```

Вы отвечаете одним из трёх чисел:  
1 — равно,  
2 — больше,  
3 — меньше.  

Напишите программу, которая с помощью цепочки таких вопросов и ответов угадывает число.  
Дополнительно: сделайте так, чтобы можно было гарантированно угадать число за семь попыток.

```python
import random
```

```python
def find_number():
    target = random.randint(1, 100)
    print(f"Загаданное число - {target}")
    top, bot = 100, 0
    
    for i in range(7):
        value = int((top + bot) / 2)
        inp = input(f"Твое число равно, меньше или больше, чем число {value}? ")
        if inp == "1":
            assert value == target, f'Неправильно, загадано {target}, решение программы {value}'
            print(f"Ваше число {value}!")
            break
        elif inp == "2":
            bot = value + 1
        elif inp == "3":
            top = value - 1

find_number()
```

### Упражнение 9.
Напишите программу, которая позволит пользователю преобразовывать числа из одной системы счисления в другую произвольным образом. Ваша программа должна поддерживать все системы счисления в диапазоне от 2 до 16 как для входных, так и для выходных данных. Если пользователь выберет систему с основанием, выходящим за границы допустимого, на экран должна быть выведена ошибка. Разделите код программы на несколько функций, включая функцию, конвертирующую число из произвольной системы счисления в десятичную, и обратную функцию, переводящую значение из десятичной системы в произвольную. В основной программе необходимо запросить у пользователя исходную систему счисления, целевую систему, а также число для преобразования.

[Калькулятор_систем_счисления](https://calcus.ru/perevod-sistem-schisleniya)

```python
def transform_digit(num: Union[str, int], notation_from: int, notation_to: int) -> int:
    if notation_from < 2 or notation_from > 16 or notation_to < 2 or notation_to > 16:
        raise ValueError("Некорректная система счисления")
    if type(num) == int:
        num = str(num)
    return from_decimal(to_decimal(num, notation_from), notation_to)
    

def to_decimal(num: str, base: int) -> int:
    mul, result = 1, 0
    for i in num[::-1]:
        if i.isdigit():
            result += (ord(i) - 48) * mul
        else:
            result += (ord(i) - 55) * mul
        mul *= base
    return result

def from_decimal(num: int, base: int) -> int:
    mul, result = 1, ""
    while num != 0:
        temp = num % base
        if temp > 9:
            result = chr(temp + 55) + result
        else:
            result = chr(temp + 48) + result
        mul *= 10
        num //= base
    return result

notation_from = int(input("Укажите исходную систему счисления: "))
notation_to = int(input("Укажите целевую систему счисления: "))
num = input("Укажите число для преобразования: ")
print(f"Итоговое число: {transform_digit(num, notation_from, notation_to)}")

# Указание правильного числа остается на совести пользователя :D (не стал вставлять проверку на это чтобы не захламлять код)
```

```python
def is_magic_date(day, month, year):
    return day * month == year % 100

def gen_magic_data(year):
    dels = get_dels(year)
    data = []
    days = {"1": 31, "2": 29 if ((year + 1900) % 4 == 0 and (year + 1900) % 100 != 0) or year % 400 == 0 else 28,
            "3": 31, "4": 30, "5": 31, "6": 30, "7": 31, "8": 31, "9": 30, "10": 31, "11": 30, "12": 31,}

    for month in dels: # Пробуем создать дату основываясь 
        if month < 13:
            day = year // month
            if 0 < day < days[str(month)]:
                data.append([day, month, year + 1900])
        else:
            break # Поскольку массивы отсортированные - можно прервать чтобы не считать лишние варианты
    return data

def get_dels(num):
    dels = []
    i = 2
    if num >= 1:
        dels.append(1) # Разделил добавление на две части чтобы не пришлось сортировать массив

    while i <= num:
        if num % i == 0:
            dels.append(i)
        i += 1
    return dels



for i in range(100):
    for j in gen_magic_data(i):
        print(f"{j[0]:2d} {j[1]:2d} {j[2]:4d}  Магическая дата - {'Да' if is_magic_date(*j) else 'Нет'}")
```
