### Задача 1. Сумма первых n положительных чисел
Напишите программу, запрашивающую у пользователя число и подсчитывающую сумму натуральных положительных чисел от 1 до введенного пользователем значения. Сумма первых n положительных чисел может быть рассчитана по формуле:

### Задача 2. Калькулятор
Составьте программу, которая запрашивает у пользователя 2 целых числа и выполняет операции:
- арифметические: `+, -, * , / , // , %, **, log10`
- сравнение: `<, <=, >, >=, !=, ==,`

выводя на экран результат каждого действия. В случае получение вещественного результата, округлите его до 2-х знаков после запятой (используя функцию round()).

````python
import math
````

````python
a, b = map(int, input("Введите числа через пробел: ").split())

print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {round(a / b, 2)}" if b != 0 else "Деление на 0 невозможно")
print(f"a // b = {a // b}" if b != 0 else "Деление на 0 невозможно")
print(f"a % b = {a % b}" if b != 0 else "Деление на 0 невозможно")
print(f"a ** b = {a ** b}")
print(f"log10(a) = {round(math.log10(a), 2)}" if a > 0 else "Взятие логарифма неотрицательного числа невозможна")
print(f"log10(b) = {round(math.log10(b), 2)}" if b > 0 else "Взятие логарифма неотрицательного числа невозможна")

print(f"a < b = {a < b}")
print(f"a <= b = {a <= b}")
print(f"a > b = {a > b}")
print(f"a >= b = {a >= b}")
print(f"a != b = {a != b}")
print(f"a == b = {a == b}")
````

### Задача 3.
Вычислите значение следующего выражения (аргументы - целые числа и вводятся с клавиатуры):
f = `((x^5 + 7) / (|-6|*y))^(1/3)/ (7 - z mod y)`
Округлите результат до 3-х знаков после запятой, используя функцию round().

````python
x, y, z = map(int, input("Введите числа x, y, z через пробелы: ").split())
try:
    f = ((x^5 + 7) / (abs(-6) * y)) ** (1/3) / (7 - z % y)
    print(f"f = {round(f, 3)}")
except:
    print("Деление на 0 невозможно")
````

### Задача 4.
Дана электрическая цепь, состоящая из 2-х последовательно соединенных проводников (сопротивление каждого известно). Найти общее сопротивление цепи (округление результата необходимо выполнить до 1-го знака после запятой).

````python
# Учитывая слова про округление до первого знака после запятой вероятнее всего имелось в виду параллельное сопротивление проводников
r1, r2 = map(int, input("Введите сопротивления проводников через пробел: ").split())
seq = r1 + r2
par = round(r1 * r2 / (r1 + r2), 1)
print(f"Сопротивление при последовательном соединении: {seq}")
print(f"Сопротивление при параллельном соединении: {par}")
````

### Задача 7. Сумма и произведение цифр в числе
Дано двузначное и трехзначное число. Для каждого выведите на экран сумму и произведение цифр.

````python
n1 = int(input("Введите двухзначное число: "))
nsum = 0
for i in str(n1):
    nsum += int(i)
nmul = 1
for i in str(n1):
    nmul *= int(i)
print(f"Сумма цифр в числе {n1}: {nsum}")
print(f"Произведение цифр в числе {n1}: {nmul}")

n2 = int(input("Введите трёхзначное число: "))
nsum = 0
for i in str(n2):
    nsum += int(i)
nmul = 1
for i in str(n2):
    nmul *= int(i)
print(f"Сумма цифр в числе {n2}: {nsum}")
print(f"Произведение цифр в числе {n2}: {nmul}")
````

### Задача 8. Сортировка трех чисел
Напишите программу, запрашивающую у пользователя три целых числа и выводящую их в упорядоченном виде – по возрастанию. Используйте функции min и max для нахождения наименьшего и наибольшего значений. Оставшееся число можно найти путем вычитания из суммы трех введенных чисел максимального и минимального.

````python
n1, n2, n3 = map(int, input("Введите три целых числа через пробел: ").split())
nmax = max(n1, n2, n3)
nmin = min(n1, n2, n3)
nmed = sum((n1, n2, n3)) - nmin - nmax
print(nmin, nmed, nmax)
````
